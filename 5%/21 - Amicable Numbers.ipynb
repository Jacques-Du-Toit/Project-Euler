```python
def sum_factors(n):
    sum_factors = 1
    for i in range(2, int(n**0.5) + 1):
        if n%i ==0:
            sum_factors += i
            sum_factors += n/i
    return int(sum_factors)
```

```python
def amicable_pairs(end):
    pairs = []
    for n in range(end):
        if n in pairs: continue
        first_sum = sum_factors(n)
        second_sum = sum_factors(first_sum)
        if first_sum!=n and second_sum==n:
            pairs.append(first_sum)
            pairs.append(second_sum)
    return sum(pairs), pairs
```

```python
amicable_pairs(10_000)
```

```python
%%timeit
amicable_pairs(10_000)
```
