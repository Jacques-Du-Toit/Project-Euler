```python
def is_prime(p):
    if p < 2:
        return False
    for i in range(2, int(p**0.5) + 1):
        if p % i == 0:
            return False
    return True

assert is_prime(1) == False
assert is_prime(2) == True
assert is_prime(3) == True
assert is_prime(5) == True
assert is_prime(6) == False
assert is_prime(21) == False
assert is_prime(23) == True
```

```python
def all_prime(str_i):
    for digit in range(len(str_i)-1):
        if not is_prime(int(str_i[digit:])):
            return False
    return True

assert all_prime('3797') == True
assert all_prime('418') == False
```
