```python
digits = {
    '0':0,
    '1':3,
    '2':3,
    '3':5,
    '4':4,
    '5':4,
    '6':3,
    '7':5,
    '8':5,
    '9':4
}
teens = {
    '10':3,
    '11':6,
    '12':6,
    '13':8,
    '14':8,
    '15':7,
    '16':7,
    '17':9,
    '18':8,
    '19':8
}
double_digits = {
    '0':0,
    '2':6,
    '3':6,
    '4':5,
    '5':5,
    '6':5,
    '7':7,
    '8':6,
    '9':6
}
```

```python
def word_num_letters(num):
    num_letters = 0
    num = str(num)
    len_num = len(num)
    if len_num > 3:
        raise ValueError('Only works for numbers 1-999')
    if len_num == 1:
        return digits[num]
    elif len_num == 3:
        num_letters += digits[num[0]] + 10
        if num[-2:] == '00':
            num_letters -= 3
            return num_letters
    if num[-2] == '1':
        num_letters += teens[num[-2:]]
    else:
        num_letters += double_digits[num[-2]]
        num_letters += digits[num[-1]]
    return num_letters
```

```python
def count_letters_to(n):
    total = 0
    for i in range(1, n+1):
        total += word_num_letters(i)
    return total
```

```python
count_letters_to(999) + 11
```

```python
%%timeit
count_letters_to(999)
```

# Convert to words

```python
digits_2 = {
    '0':'',
    '1':'one',
    '2':'two',
    '3':'three',
    '4':'four',
    '5':'five',
    '6':'six',
    '7':'seven',
    '8':'eight',
    '9':'nine'
}
teens_2 = {
    '10':'ten',
    '11':'eleven',
    '12':'twelve',
    '13':'thirteen',
    '14':'fourteen',
    '15':'fifteen',
    '16':'sixteen',
    '17':'seventeen',
    '18':'eighteen',
    '19':'nineteen'
}
double_digits_2 = {
    '0':'',
    '2':'twenty',
    '3':'thirty',
    '4':'forty',
    '5':'fifty',
    '6':'sixty',
    '7':'seventy',
    '8':'eighty',
    '9':'ninety'
}
```

```python
def num_to_word(n):
    word = ''
    num = str(n)
    len_num = len(num)
    if len_num > 3: raise ValueError('Too big')
    elif len_num == 1: return digits_2[num]
    elif len_num ==3:
        word += digits_2[num[-3]]
        word += ' hundred '
        if num[-2:] != '00':
            word += 'and '
    if num[-2] == '1':
        word += teens_2[num[-2:]]
    else:
        word += double_digits_2[num[-2]]
        word += digits_2[num[-1]]
    return word
```

```python
def count_word_lens(n):
    t = 0
    for i in range(1, n+1):
        word = num_to_word(i)
        t+=len(num_to_word(i).replace(' ', ''))
    return t
```

```python
%%timeit
count_word_lens(999)
```
