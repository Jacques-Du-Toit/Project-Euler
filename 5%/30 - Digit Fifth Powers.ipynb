The bigger the number is the larger it gets in relation to the sum of its digits and their powers (it gets roughly \*10 where as they only ever add a finite amount).  
The number grows at $10^{d-1}$ where as the sum of digits grows linearly with $d$ where $d$ is the number of digits in the number.  
The largest a set of digits can be to the nth power is $9^{n}+...+9^{n} = d*9^{n}$  
So when $10^{d-1} > d*9^{n}$, then the digits can no longer ever catch up and we have our upper bound.  
  
Say we are working with $n = 4$, this means the largest the sum of digits can be is $d*9^{4} = d*6561$  
$10^{4-1}=10000 \leq 26244 = 4*6561$ but $10^{5-1}=100000 > 32805 = 5*6561$ so we know we can only have .  

```python
def nth_power_digits(number, n):
    total = 0
    num = str(number)
    for dig in num:
        total += int(dig)**n
    if total == number:
        return True
    else:
        return False
```
