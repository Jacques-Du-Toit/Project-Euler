```python
mapping = {'T':10, 'J':11, 'Q':12, 'K':13, 'A':14}
```

```python
def is_straight(vals):
    if len(set(vals)) < 5:
        return False
    if vals == [v for v in range(min(vals), max(vals)+1)]:
        return True
    else:
        return False

assert is_straight([3, 4, 5, 6, 7]) == True
assert is_straight([3, 4, 6, 7]) == False
assert is_straight([3, 4, 5, 6, 8]) == False
assert is_straight([2, 4, 5, 6, 7]) == False
assert is_straight([8, 9, 10, 11, 12]) == True
```

```python
def rank_hand(hand):
    straight = False
    flush = False
    # Get the values and suits
    vals = [c[:-1] for c in hand]
    vals = [
        mapping[val] if val in mapping else int(val) for val in vals
    ]
    vals.sort()
    suits = {c[-1] for c in hand}
    # Get high card
    high = max(vals)
    
    if is_straight(vals):
        straight = True

    if len(suits) == 1:
        flush = True
    # Straight Flush
    if straight and flush:
        return [9, high]

    unique_counts = {
        val:vals.count(val) for val in set(vals)
    }
    # 4 of a kind
    if 4 in unique_counts.values():
        four = [key for (key, val) in unique_counts.items() if val == 4][0]
        left = [x for x in vals if x!=four][0]
        return [8, four, left]
        
    is_three = False
    is_two = False
    if 3 in unique_counts.values():
        is_three = True
    if 2 in unique_counts.values():
        is_two = True

    if is_three and is_two:
        three = [key for (key, val) in unique_counts.items() if val == 3][0]
        two = [key for (key, val) in unique_counts.items() if val == 2][0]
        return [7, three, two]

    if flush:
        return [6] + vals[::-1]

    if straight:
        return [5, high]

    if is_three:
        three = [key for (key, val) in unique_counts.items() if val == 3][0]
        left = [x for x in vals if x!=three]
        return [4, three] + left[::-1]

    if is_two:
        two = [key for (key, val) in unique_counts.items() if val == 2]
        left = [x for x in vals if x not in two]
        if len(two) == 2:
            two.sort()
            return [3] + two[::-1] + left
        return [2, two[0]] + left[::-1]

    return [1] + vals[::-1]

assert rank_hand(['QS', '9S', 'TS', '8S', 'JS']) == [9, 12]
assert rank_hand(['TC', '9S', 'TS', 'TD', 'TH']) == [8, 10, 9]
assert rank_hand(['TC', '9S', 'TS', '9D', '9H']) == [7, 9, 10]
assert rank_hand(['QS', '9S', '4S', '8S', 'JS']) == [6, 12, 11, 9, 8, 4]
assert rank_hand(['QS', '4S', '4S', '8S', 'JS']) == [6, 12, 11, 8, 4, 4]
assert rank_hand(['QS', '9D', 'TS', '8S', 'JS']) == [5, 12]
assert rank_hand(['QS', 'QD', 'QS', '8S', 'JS']) == [4, 12, 11, 8]
assert rank_hand(['AS', 'AD', '8S', '8S', 'JS']) == [3, 14, 8, 11]
assert rank_hand(['3S', '2D', '8S', '8S', 'JS']) == [2, 8, 11, 3, 2]
assert rank_hand(['3S', '2D', '8S', '7S', 'JS']) == [1, 11, 8, 7, 3, 2]
```

```python
hands['1 Score'] = hands['Player 1'].apply(rank_hand)
hands['2 Score'] = hands['Player 2'].apply(rank_hand)
```

```python
def winner(one, two):
    for score_one, score_two in zip(one, two):
        if score_one < score_two:
            return '2'
        elif score_one > score_two:
            return '1'

assert winner([1, 13, 10, 9, 8, 4], [1, 14, 7, 5, 3, 2]) == '2'
```
