Obviously there should be a way to rule out all pairs that are going to be equal at the start through an equation

# Brute Force

```python
# Using Set
def find_with_set():
    terms = {0}
    for a in range(2, 101):
        for b in range(2, 101):
            terms.add(a**b)
    
    return len(terms)-1
```

```python
%%timeit
find_with_set()
```

```python
# Using List
def find_with_list_1():
    terms = []
    for a in range(2, 101):
        for b in range(2, 101):
            term = a**b
            if term not in terms:
                terms.append(term)
    
    return len(terms)


def find_with_list_2():
    terms = []
    for a in range(2, 101):
        for b in range(2, 101):
            terms.append(a**b)
    
    return len(set(terms))
```

```python
%%timeit
find_with_list_1()
```

```python
%%timeit
find_with_list_2()
```
