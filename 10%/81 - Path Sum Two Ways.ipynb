Maybe an easier way of programming this would be to begin with the last row (and last column) working your way left, add them across, then the next last row and so on

Lets now make an algorithm to do this for us

```python
def find_min_sum(matrix_str, seperator):
    # Get into right format
    matrix = [row.split(seperator) for row in matrix_str.split('\n') if row!='']
    matrix = [[int(val) for val in row] for row in matrix]
    # Sum the last row (from the last column backwards)
    matrix[-1] = [sum(matrix[-1][-i:]) for i in range(len(matrix[-1]), 0, -1)]
    # Transpose matrix
    matrix = [row[::-1] for row in matrix[::-1]]
    # Work your way from the bottom row up
    for row_idx, row in enumerate(matrix[1:]):
        r_idx = row_idx+1
        for v_idx, val in enumerate(row):
            # Check if the left index is within bounds, then assign 'left' accordingly.
            left = matrix[r_idx][v_idx-1] if v_idx > 0 else np.inf
            
            above = matrix[r_idx-1][v_idx] if r_idx > 0 else np.inf
    
            matrix[r_idx][v_idx] += min(left, above)
    return matrix[-1][-1]
```
