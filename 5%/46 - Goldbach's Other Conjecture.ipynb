```python
def is_prime(p):
    for pos_fac in range(2, int(p**0.5)+1):
        if p%pos_fac==0:
            return False
    return True

assert is_prime(2) == True
assert is_prime(3) == True
assert is_prime(4) == False
assert is_prime(5) == True
assert is_prime(6) == False
assert is_prime(7) == True
assert is_prime(9) == False
assert is_prime(21) == False
assert is_prime(29) == True
```

```python
def can_be_written(comp):
    num = 1
    square = 2*num**2
    while square < comp:
        left = comp - square
        if is_prime(left):
            return True
        num += 1
        square = 2*num**2
    return False

assert can_be_written(9) == True
assert can_be_written(15) == True
assert can_be_written(21) == True
assert can_be_written(25) == True
assert can_be_written(27) == True
```
