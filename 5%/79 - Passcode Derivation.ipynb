```python
def find_pass(attempts_str):
    attempts = {a for a in attempts_str.split('\n') if a!=''}
    passcode = []
    while True:
        current = {a[0] for a in attempts}
        current = [num for num in current if len([a for a in attempts if a[0]!=num and num in a])==0]
        if len(current) != 1:
            raise ValueError(f'{current} values for next value')
        current = current[0]
        passcode.append(current)
        attempts = {a.replace(current, '') for a in attempts if a!=current}
        if len(attempts) == 1:
            passcode.append(list(attempts)[0])
            return ''.join(passcode)
```
