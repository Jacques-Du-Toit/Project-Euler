```python
def is_palindrome(num):
    if str(num) == str(num)[::-1]:
        return True
    else:
        return False

assert is_palindrome(585) == True
assert is_palindrome(1001001001) == True
assert is_palindrome(125213) == False
```

```python
def highest_power_of_2(num):
    if num == 0:
        return 0, 0
    elif num == 1:
        return 1, 0
        
    i = 0
    while True:
        next = 2**i
        # If num bigger we need to increase next until we get the next biggest
        if next < num:
            i+=1
        # If they are equal, we have an exact power of 2 with no remainder
        elif num == next:
            return 0, i
        # If num smaller than next, the previous i with the bigger power below
            # and we can return that i
        else:
            prev = next//2
            return num - prev, i-1

assert highest_power_of_2(0) == (0, 0)
assert highest_power_of_2(1) == (1, 0)
assert highest_power_of_2(2) == (0, 1)
assert highest_power_of_2(3) == (1, 1)
assert highest_power_of_2(4) == (0, 2)
assert highest_power_of_2(16) == (0, 4)
assert highest_power_of_2(17) == (1, 4)
```

```python
def int_to_binary(num):
    if num == 0:
        return 0
        
    
    remainder, power = highest_power_of_2(num)
    binary = ['0']*(power+1)
    binary[power] = '1'
    while True:
        if remainder == 0 or power == 0:
            return int(''.join(binary)[::-1])
            
        remainder, power = highest_power_of_2(remainder)
        binary[power] = '1'

    

assert int_to_binary(0) == 0
assert int_to_binary(1) == 1
assert int_to_binary(2) == 10
assert int_to_binary(3) == 11
assert int_to_binary(4) == 100
assert int_to_binary(585) == 1001001001
```

```python
total = 0
for i in range(1, 1_000_000):
    if is_palindrome(i) and is_palindrome(int_to_binary(i)):
        total+=i
    print(f'{round((i+1)/10_000, 2)}%', end='\r')
```
