```python
def factorial(n):
    total = 1
    for i in range(1, n+1):
        total *= i
    return total
```

```python
def find_next_start(ordered_digits, place):
    current_place = place-1
    digits_left = len(ordered_digits)-1
    digit_order = []
    while digits_left >= 0:
        index = current_place//factorial(digits_left)
        digit = ordered_digits[index]
        digit_order.append(str(digit))
        ordered_digits.remove(digit)
        current_place -= factorial(digits_left)*index
        digits_left -= 1
    return ''.join(digit_order)
```

```python
find_next_start([x for x in range(10)], 1_000_000)
```

```python
%%timeit
find_next_start([x for x in range(10)], 1_000_000)
```
