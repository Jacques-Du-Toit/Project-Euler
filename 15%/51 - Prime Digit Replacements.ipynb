Does the replacement need to be on the same digits i.e. does 71341 need 1 and 1 replaced at the same time by the same digit or could it be 1 and 3?  
Im going to assume yes

- Going from 10 onwards (digits will just give, check if number is prime, if it is, try replacing 1 digit in it with each other number (maybe not last with anything that contains 2, 4, 5, 6, 8, 0 as this will not ever be prime)
- then check how many of those are prime
- move the 1 digit to the next place repeat
- then try 2 digits .. until you have the number of digits the number is long

```python
def is_prime(p):
    if p == 1 or p == 2:
        return True
    elif p < 1:
        return False
    for i in range(2, int(p**0.5) + 1):
        if p%i == 0:
            return False
    return True
```

```python
def replace_digits(prime, family_size):
    """
    Counts the size of the family of primes for each prime digit combination
    We dont keep track of the family as we only care once it hits a size of 8,
    the number we are testing would be the smallest of those 8 as it came first
    """
    sp = str(prime)
    num_range = [str(n) for n in range(0, 10)]
    for rep_digit in num_range:
        family = 0
        for new_digit in num_range:
            # Cant have 0 at the start and change the digit size (or 13 would have 7 members)
            if rep_digit == sp[0] and new_digit == '0':
                continue
            new = int(sp.replace(rep_digit, new_digit))
            if is_prime(new):
                family += 1
        if family == family_size:
            return prime, rep_digit
    return 0, ''
```

```python
def find_smallest_family(family_size=8):
    num = 10
    while True:
        if is_prime(num):
            prime, dig = replace_digits(num, family_size)
            if prime != 0:
                return prime, dig
        num += 1
```
