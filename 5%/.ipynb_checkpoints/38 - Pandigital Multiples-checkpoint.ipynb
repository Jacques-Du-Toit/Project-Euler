$n > 1$ so cant just choose $987654321$,  
We know $n \neq 9$ as if we choose $p > 1$, would have more than 9 digits, $p = 1$ gives nine 1's.  
Same for $n \neq 8, 7, 6$ as if we choose $p > 1$, would have more than 9 digits  

```python
def find_largest_p():
    p = 1
    largest = 0
    # See if possible for this n and p
    while True:
        total = ''
        for n in range(1, 10):
            total += str(p*n)
            if '0' in total:
                break
            len_total = len(total)
            if len({a for a in total}) == 9 and len_total ==9:
                largest = total
            
            if len_total > 9:
                if n == 2:
                    return largest
                break
        p+=1
```

```python
find_largest_p()
```
