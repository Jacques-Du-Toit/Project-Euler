```python
from time import time
import matplotlib.pyplot as plt
from math import sqrt
```

```python
def plot_different_function_times(functions, function_names, parameters, scale='linear'):
    time_lists = [[] for func in range(len(functions))]
    for param in parameters:
        for function, time_list in zip(functions, time_lists):   
            times = []
            for repeat in range(5):
                start = time()
                function(param)
                taken = time() - start
                times.append(taken)
            time_list.append(sum(times)/len(times))

    for i, name in enumerate(function_names):
        plt.plot(parameters, time_lists[i], label=name)
    plt.legend()
    plt.xlabel('param')
    plt.ylabel('Average Time (s)')
    if scale == 'log':
        plt.yscale('log')
    plt.show()
```

```python
def basic_function(b, r):
    first = b/(b+r)
    second = (b-1)/(b+r-1)
    if abs(first*second - 0.5) <= 0.00000000001:
        return True
    else:
        return False

assert basic_function(15, 6) == True
assert basic_function(85, 35) == True
assert basic_function(42, 22) == False
```

```python
def second_function(b, r):
    if 2*b*(b-1) == (b+r)*(b+r-1):
        return True
    else:
        return False

assert second_function(15, 6) == True
assert second_function(85, 35) == True
assert second_function(42, 22) == False
```

```python
def brute_force_answer(start_discs, function):
    n = start_discs + 1
    while True:
        for b in range(1, n):
            if function(b, n-b):
                return n, b, n-b
        n+=1

assert brute_force_answer(10, basic_function) == (21, 15, 6)
assert brute_force_answer(21, basic_function) == (120, 85, 35)
assert brute_force_answer(10, second_function) == (21, 15, 6)
assert brute_force_answer(21, second_function) == (120, 85, 35)
```

```python
def next_ones_brute(function):
    n = 1
    found_ns = []
    found_bs = []
    found_rs = []
    while True:
        for b in range(1, n):
            if function(b, n-b):
                found_ns.append(n)
                found_bs.append(b)
                found_rs.append(n-b)
                print(n, b, n-b)
        n+=1
        print(n, end='\r')
        if n == 24000:
            return found_ns, found_bs, found_rs
```

This is making a lot of sense, as b and b+r tend to infinity, the difference between $\frac{b}{b+r}$ and $\frac{b-1}{b+r-1}$ will become more and more negligible, so at infinity we can think of these as the same fraction,  
so the fraction $f = \sqrt{\frac{1}{2}}$ so that $f*f = \frac{1}{2}$ => $f = \frac{1}{\sqrt{2}}$  
We can see the ratio of $\frac{b}{b+r}$ tending to this  
Maybe this is to do with getting the next nearest fractions above and below $\frac{1}{\sqrt{2}}$  
$\frac{b}{b+r} > \frac{b-1}{b+r-1}$ so $\frac{b}{b+r} > \frac{1}{\sqrt{2}}$ and $\frac{b-1}{b+r-1} < \frac{1}{\sqrt{2}}$  

```python
# b will always be above half of n
def brute_force_answer_from_half(start_discs):
    n = start_discs + 1
    while True:
        for b in range(n//2+1, n):
            if second_function(b, n-b):
                return n, b, n-b
        n+=1

assert brute_force_answer_from_half(10) == (21, 15, 6)
assert brute_force_answer_from_half(21) == (120, 85, 35)
assert brute_force_answer_from_half(700) == (4060, 2871, 1189)

# b will always be above n/root(2)
def brute_force_answer_from_root_two(start_discs):
    sqrt_two = sqrt(2)
    n = start_discs + 1
    while True:
        for b in range(int(n/sqrt_two)+1, n):
            if second_function(b, n-b):
                return n, b, n-b
        n+=1

assert brute_force_answer_from_root_two(10) == (21, 15, 6)
assert brute_force_answer_from_root_two(21) == (120, 85, 35)
assert brute_force_answer_from_root_two(700) == (4060, 2871, 1189)

def next_ones_brute_from_root_two(stop=24000):
    n = 1
    sqrt_two = sqrt(2)
    found_ns = []
    found_bs = []
    found_rs = []
    while True:
        for b in range(int(n/sqrt_two)+1, n):
            if second_function(b, n-b):
                found_ns.append(n)
                found_bs.append(b)
                found_rs.append(n-b)
                print(n, b, n-b)
        n+=1
        print(n, end='\r')
        if n == stop:
            return found_ns, found_bs, found_rs

def worst(start_discs):
    n = start_discs + 1
    while True:
        for b in range(1, n):
            if basic_function(b, n-b):
                return n, b, n-b
        n+=1
```

```python
def resilient(numerator, denominator):
    if numerator == 1:
        return True
    if denominator%numerator == 0:
        return False
    # Check what are the factors of numerator, see if denominator shares any
    for possible_divisor in range(2, int(sqrt(numerator))+1):
        if numerator%possible_divisor == 0:
            if denominator%possible_divisor == 0:
                return False
            if denominator%(numerator//possible_divisor) == 0:
                return False
    # If they share no factors the fraction is resilient (there exists no a)
    return True
```

```python
def simplify(numerator, denominator, cf=1):
    if numerator == 1:
        return numerator, denominator, cf
    if denominator%numerator == 0:
        return 1, denominator//numerator, numerator
    # Check what are the factors of numerator, see if denominator shares any
    for possible_divisor in range(2, int(sqrt(numerator))+1):
        if numerator%possible_divisor == 0:
            other_divisor = numerator//possible_divisor
            if denominator%other_divisor == 0:
                new_num, new_den, new_cf = simplify(
                    numerator//other_divisor,
                    denominator//other_divisor
                )
                return new_num, new_den, cf*new_cf*other_divisor
            elif denominator%possible_divisor == 0:
                new_num, new_den, new_cf =  simplify(
                    numerator//possible_divisor,
                    denominator//possible_divisor
                )
                return new_num, new_den, cf*new_cf*possible_divisor
                
    # If they share no factors the fraction is resilient (there exists no a)
    return numerator, denominator, cf

assert simplify(1, 2) == (1, 2, 1)
assert simplify(3, 15) == (1, 5, 3)
assert simplify(2, 3) == (2, 3, 1)
assert simplify(15, 165) == (1, 11, 15)
assert simplify(15, 55) == (3, 11, 5)
```
