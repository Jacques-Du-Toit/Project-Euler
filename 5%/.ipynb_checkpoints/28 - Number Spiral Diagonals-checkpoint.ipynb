Can be seen that for every layer ther spiral goes out, the distance between diagonals increases by 2, this is because the size of the next layer it 2 wider, when transitioning between layers, one move is required to move right out of first layer, but a move is also saved because it starts lower, so rule not broken.  

# Brute Force

```python
def count_as_we_go(side_length):
    layers = (side_length+1)//2
    total = 1
    current = 1
    for layer in range(1, layers):
        for i in range(4):
            current += 2*layer
            total += current
    return total
```

```python
count_as_we_go(1001)
```

```python
%%timeit
count_as_we_go(1001)
```

# Equation

```python
def equation(side_length):
    layers = (side_length+1)//2
    total = 1 + 4*(layers-1)
    for layer in range(2, layers+1):
        total+= 16*layer**2 - 28*layer + 12
    return total
```

```python
equation(1001)
```

```python
%%timeit
equation(1001)
```
