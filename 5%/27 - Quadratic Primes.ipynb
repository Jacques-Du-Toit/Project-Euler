We need to check if a number is prime, then move to the next value of n (for a given a and b) until we dont hit a prime, return the number of consecutive primes

```python
def is_prime(p):
    if p == 1 or p == 2:
        return True
    elif p < 1:
        return False
    for i in range(2, int(p**0.5) + 1):
        if p%i == 0:
            return False
    return True
```

```python
assert is_prime(2) == True
assert is_prime(5) == True
assert is_prime(6) == False
assert is_prime(11) == True
assert is_prime(21) == False
```

```python
def consecutive_primes(a, b):
    total = 0
    n = 0
    while True:
        if is_prime(n**2 + (a*n) + b):
            n+=1
            total+=1
        else:
            return total
```

```python
assert consecutive_primes(1, 41) == 40
assert consecutive_primes(-79, 1601) == 80
```

```python
def max_primes_coef(a_range = 1000, b_range = 1001):
    most = 0
    a_best = 0
    b_best = 0
    for a in range(-1*a_range + 1, a_range):
        for b in range(-1*b_range + 1, b_range):
            next = consecutive_primes(a, b)
            if next > most:
                most = next
                a_best = a
                b_best = b
    return a_best, b_best, a_best*b_best, most
```
