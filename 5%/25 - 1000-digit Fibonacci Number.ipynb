Comparing length of string is much slower than size of int

```python
def first_fib_n_digits(n):
    if n == 1: return 1, 0
    first = 1
    second = 1
    index = 3
    upper = 10**(n-1)
    while True:
        dummy = first + second
        if dummy >= upper:
            return dummy, index
        first = second
        second = dummy
        index+=1
```

```python
first_fib_n_digits(1000)
```

```python
%%timeit
first_fib_n_digits(1000)
```
