### Explanations

- I'm creating lots of sub lists per row of the matrix
- Then going to go number by number, checking the multiples of different directions
- If you check left, then no need to check right as youd be checking the same multiple twice in a row
- Same goes for up and down
- I need to atleast check both diagonal though, only once each, so not up left and down right, as they are on the same diagonal

```python
# Use these to check if we should even check in these directions for a multiple
def left_eligible(col_num, look_range):
    if col_num+1 - look_range >= 0:
        return True
    else:
        return False


def right_eligible(col_num, look_range, len_row):
    if col_num+look_range <= len_row:
        return True
    else:
        return False


def up_eligible(row_num, look_range):
    if row_num+1 - look_range >= 0:
        return True
    else:
        return False


def left_calc(num, row, col_num, look_range, highest_multiple, highest_multiples):
    multiples = [num]
    multiple = num
    for step in range(1, look_range):
        val = row[col_num-step]
        multiples.append(val)
        multiple *= val
    if multiple > highest_multiple:
        highest_multiple = multiple
        highest_multiples = multiples
    return highest_multiple, highest_multiples


def up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples):
    multiples = [num]
    multiple = num
    for step in range(1, look_range):
        val = rows[row_num-step][col_num]
        multiples.append(val)
        multiple *= val
    if multiple > highest_multiple:
        highest_multiple = multiple
        highest_multiples = multiples
    return highest_multiple, highest_multiples


def left_up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples):
    multiples = [num]
    multiple = num
    for step in range(1, look_range):
        val = rows[row_num-step][col_num-step]
        multiples.append(val)
        multiple *= val
    if multiple > highest_multiple:
        highest_multiple = multiple
        highest_multiples = multiples
    return highest_multiple, highest_multiples


def right_up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples):
    multiples = [num]
    multiple = num
    for step in range(1, look_range):
        val = rows[row_num-step][col_num+step]
        multiples.append(val)
        multiple *= val
    if multiple > highest_multiple:
        highest_multiple = multiple
        highest_multiples = multiples
    return highest_multiple, highest_multiples


# Main Function
def best_multiple(grid, len_row, look_range):
    highest_multiple = 0
    highest_multiples = []
    # Make individual lists per row in the grid, rows contains each of these
    rows = [grid[i:i+len_row] for i in range(0, len(grid), len_row)]
    for row_num, row in enumerate(rows):
        for col_num, num in enumerate(row):
            left_el = left_eligible(col_num, look_range)
            right_el = right_eligible(col_num, look_range, len_row)
            up_el = up_eligible(row_num, look_range)
            if left_el:
                highest_multiple, highest_multiples = left_calc(num, row, col_num, look_range, highest_multiple, highest_multiples)
                if up_el:
                    highest_multiple, highest_multiples = left_up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples)
            if up_el:
                highest_multiple, highest_multiples = up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples)
                if right_el:
                    highest_multiple, highest_multiples = right_up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples)
    return highest_multiple, highest_multiples
```

```python
%%ipytest

@pytest.mark.parametrize('col_num, look_range, eligible',[
    (2, 3, True),
    (1, 3, False),
    (2, 4, False),
])
def test_left_eligible(col_num, look_range, eligible):
    assert left_eligible(col_num, look_range) == eligible

@pytest.mark.parametrize('col_num, look_range, len_row, eligible',[
    (1, 2, 3, True),
    (5, 10, 30, True),
    (1, 2, 2, False),
    (1, 3, 3, False),
    (2, 2, 3, False),
])
def test_right_eligible(col_num, look_range, len_row, eligible):
    assert right_eligible(col_num, look_range, len_row) == eligible

@pytest.mark.parametrize('row_num, look_range, eligible',[
    (1, 2, True),
    (1, 3, False),
    (0, 2, False),
    (2, 5, False),
    (10, 9, True)
])
def test_up_eligible(row_num, look_range, eligible):
    assert up_eligible(row_num, look_range) == eligible

@pytest.mark.parametrize('num, row, col_num, look_range, highest_multiple, highest_multiples, new_highest_multiple, new_highest_multiples',[
    (
        5, [60, 22, 5], 2, 2, 0, [],
        110,  [5, 22]
    ),
    (
        18, [17, 3, 7, 18, 90, -5], 3, 3, 0, [],
        378,  [18, 7, 3]
    )
])
def test_left_calc(num, row, col_num, look_range, highest_multiple, highest_multiples, new_highest_multiple, new_highest_multiples):
    assert left_calc(num, row, col_num, look_range, highest_multiple, highest_multiples) == (new_highest_multiple, new_highest_multiples)

@pytest.mark.parametrize('num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples, new_highest_multiple, new_highest_multiples',[
    (
        2, [[1, 5, 2, 4], [1, 4, 9, 0], [12, 2, 7, -1], [8, 2, 6, 5]],
        2, 1, 2, 0, [],
        8, [2, 4]
    )
])
def test_up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples, new_highest_multiple, new_highest_multiples):
    assert up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples) == (new_highest_multiple, new_highest_multiples)

@pytest.mark.parametrize('num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples, new_highest_multiple, new_highest_multiples',[
    (
        7, [[1, 5, 2, 4], [1, 4, 9, 0], [12, 2, 7, -1], [8, 2, 6, 5]],
        2, 2, 3, 0, [],
        28, [7, 4, 1]
    )
])
def test_left_up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples, new_highest_multiple, new_highest_multiples):
    assert left_up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples) == (new_highest_multiple, new_highest_multiples)

@pytest.mark.parametrize('num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples, new_highest_multiple, new_highest_multiples',[
    (
        8, [[1, 5, 2, 4], [1, 4, 9, 0], [12, 2, 7, -1], [8, 2, 6, 5]],
        3, 0, 4, 0, [],
        576, [8, 2, 9, 4]
    )
])
def test_right_up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples, new_highest_multiple, new_highest_multiples):
    assert right_up_calc(num, rows, row_num, col_num, look_range, highest_multiple, highest_multiples) == (new_highest_multiple, new_highest_multiples)

test_grid_3 = [
    1, 22, 5,
    12, -6, 5,
    0, 18, 4
]

test_grid_4 = [
    1, 5, 2, 4,
    1, 4, 9, 0,
    12, 2, 7, -1,
    8, 2, 6, 5
]

@pytest.mark.parametrize('grid, len_row, look_range, highest_multiple, highest_multiples',[
    (
        test_grid_3, 3, 2,
        264, [12, 22]
    ),
    (
        test_grid_4, 4, 3,
        378, [6, 7, 9]
    )
])
def test_best_multiple(grid, len_row, look_range, highest_multiple, highest_multiples):
    assert best_multiple(grid, len_row, look_range) == (highest_multiple, highest_multiples)
```
