Trying instead for $\frac{b}{d} * \frac{b-1}{d-1} = \frac{1}{4}$

```python
def follows_rule_4(d):
    b = int(d//2)+1
    if 4*b*(b-1) == d*(d-1):
        return True
    else:
        return False
```

```python
d = 2
while True:
    if follows_rule_4(d):
        print(b, d, d-b, b/d, (d-b)/d)
    d+=1
```

Would actually be faster to search b instead of d space

Trying instead for $\frac{b}{d} * \frac{b-1}{d-1} * \frac{b-2}{d-2} = \frac{1}{2}$

```python
def follows_rule_3_2s(d, b):
    if 2*b*(b-1)*(b-2) == d*(d-1)*(d-2):
        return True
    else:
        return False
```

```python
b = 4
cube_root_two = 2**(1/3)
while True:
    for subtract in range(-1, 4):
        d = int(cube_root_two*b) + subtract
        print(f'{b}/{d} * {b-1}/{d-1} * {b-2}/{d-2} == {b/d * (b-1)/(d-1) * (b-2)/(d-2)}')
        if follows_rule_3_2s(d, b):
            print(b, d)
    b+=1
    if b == 100:
        break
```
