Will just a attempt a brute force method

```python
def current_coin(coins, coin_i, max, running_total):
    total = 0
    coin = coins[coin_i]

    if coin == coins[-1]:
        if (max - running_total)%coin==0: 
            return 1
        else:
            return 0
    
    for num_coin in range(max//coin, -1, -1):
        cur_run = running_total + num_coin*coin
        if cur_run >= max:
            if cur_run == max:
                total+=1
        else:
            total += current_coin(coins, coin_i+1, max, cur_run)
    return total
```

```python
current_coin([200, 100, 50], 0, 200, 0)
```

```python
def find_totals(coins_old=[1, 2, 5, 10, 20, 50, 100, 200], max=200):
    coins = coins_old.copy()
    coins.sort()
    coins = coins[::-1]
    return current_coin(coins, 0, max, 0)
```
