```python
def days_in_feb(year):
    if year%400==0:
        return 29
    elif year%100==0:
        return 28
    elif year&4==0:
        return 29
    else:
        return 28
```

```python
def num_sundays(start, end):
    total_sundays = 0
    current_day = 1
    for year in range(1900, end):
        days_in_month = [
            31, days_in_feb(year), 31 ,30, 31, 30, 31, 31, 30, 31, 30, 31
        ]
        for month_days in days_in_month:
            if current_day == 0 and year>=start:
                total_sundays += 1
            current_day = (current_day + month_days)%7
    return total_sundays
```

```python
num_sundays(1901, 2001)
```

```python
%%timeit
num_sundays(1901, 2001)
```
