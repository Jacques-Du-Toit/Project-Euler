```python
def reverse_num(num):
    return int(str(num)[::-1])

def is_lychrel(num):
    forwards = num
    for i in range(49):
        backwards = reverse_num(forwards)
        next = forwards + backwards
        if next == reverse_num(next):
            return False
        forwards = next
    return True
```

```python
is_lychrel(47)
```

```python
def how_many_lychrel_below(n):
    total = 0
    for i in range(1, n):
        if is_lychrel(i):
            total += 1
    return total
```

```python
how_many_lychrel_below(10000)
```
