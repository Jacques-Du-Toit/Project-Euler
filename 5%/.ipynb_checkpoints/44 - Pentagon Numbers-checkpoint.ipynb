## Problem Formulation
We have these constraints:  
1. $j(3j - 1) + k(3k - 1) = x(3x - 1)$ *(their sum is a pentagon number)*
2. $j(3j - 1) - k(3k - 1) = y(3y - 1)$ *(their difference is a pentagon number)*
3. $j, k, x, y \in \mathbb{N}, x > j > k > y$

# Brute Force

Generate a new pentagon number, see if its sum and diff with any of previous pentagon numbers generates two pentagon numbers.  
The hard part is ensuring that the difference will also be a pentagon number, as we could theoretically have two extremely large pentagon numbers that's difference is smaller than the initial pairing we find.  
e.g. if we find that $P_1, P_{1000}$ works - we might then discover that $P_{1000}, P_{1001}$ also works but with a smaller difference.
